# Lección 1: Fundamentos de Big Data

## Introducción

En la actualidad, el crecimiento exponencial de los datos ha transformado la forma en que las organizaciones toman decisiones. Cada interacción digital —como compras en línea, uso de redes sociales o sensores IoT— genera grandes volúmenes de información que deben ser procesados y analizados.

En este contexto surge el concepto de **Big Data**, el cual hace referencia al manejo de grandes volúmenes de datos que no pueden ser procesados eficientemente mediante herramientas tradicionales.

El objetivo de esta lección es comprender los fundamentos de Big Data, sus características principales y su ecosistema tecnológico.



## ¿Qué es Big Data?

Big Data se refiere al conjunto de tecnologías, procesos y herramientas diseñadas para gestionar, almacenar y analizar grandes volúmenes de datos, tanto estructurados como no estructurados.

A diferencia de los sistemas tradicionales, Big Data permite:
- Procesar datos a gran escala
- Analizar información en tiempo real
- Extraer valor para la toma de decisiones



## Las 5V del Big Data

El concepto de Big Data se define mediante cinco dimensiones clave:

### 1. Volumen
Se refiere a la enorme cantidad de datos generados continuamente, desde gigabytes hasta petabytes o más.

### 2. Velocidad
Hace referencia a la rapidez con la que los datos son generados, procesados y analizados.

### 3. Variedad
Incluye la diversidad de formatos de datos:
- Estructurados (bases de datos)
- No estructurados (texto, imágenes, video)
- Semiestructurados (JSON, XML)

### 4. Veracidad
Se relaciona con la calidad y confiabilidad de los datos.

### 5. Valor
El objetivo final: transformar datos en información útil para la toma de decisiones.



## Problemas que resuelve Big Data

Big Data permite abordar desafíos que los sistemas tradicionales no pueden resolver, como:

- Procesamiento de grandes volúmenes de datos
- Análisis en tiempo real
- Detección de fraudes
- Predicción de comportamiento
- Optimización de procesos



## Sistemas locales vs distribuidos

### Sistema local
- Procesamiento en una sola máquina
- Limitado en capacidad y escalabilidad

### Sistema distribuido
- Procesamiento en múltiples nodos
- Alta escalabilidad
- Mayor tolerancia a fallos

 Big Data se basa principalmente en sistemas distribuidos.



## Ecosistema Big Data

El ecosistema Big Data incluye diversas tecnologías:

- **Almacenamiento distribuido**: Hadoop (HDFS)
- **Procesamiento distribuido**: Apache Spark
- **Bases de datos NoSQL**: MongoDB, Cassandra
- **Procesamiento en tiempo real**: Kafka, Flink

Estas herramientas trabajan en conjunto para gestionar el ciclo completo de los datos.



## Arquitectura típica de Big Data

Una arquitectura Big Data se compone de:

1. **Ingesta de datos**

# Lección 3: RDD Avanzado

## Introducción

En esta sección se profundiza en el uso de RDDs mediante transformaciones más avanzadas, incluyendo el uso de estructuras clave-valor (PairRDD) y operaciones de agregación distribuidas como reduceByKey.

Estas herramientas permiten replicar el modelo MapReduce, base del procesamiento de grandes volúmenes de datos en entornos distribuidos.

```python
datos = ["hola mundo", "spark es genial"]

rdd_texto = sc.parallelize(datos)

# map → mantiene estructura
map_result = rdd_texto.map(lambda x: x.split(" "))
print("map:", map_result.collect())

# flatMap → aplana estructura
flatmap_result = rdd_texto.flatMap(lambda x: x.split(" "))
print("flatMap:", flatmap_result.collect())
```

```python
# PairRDD (clave-valor)

rdd_palabras = rdd_texto.flatMap(lambda x: x.split(" "))

# Crear pares (palabra, 1)
rdd_pares = rdd_palabras.map(lambda palabra: (palabra, 1))

print(rdd_pares.collect())
```

```python
# reduceByKey

conteo = rdd_pares.reduceByKey(lambda a, b: a + b)

print(conteo.collect())
```

```python
texto = [
    "big data es poderoso",
    "spark procesa big data",
    "big data es el futuro"
]

rdd_texto2 = sc.parallelize(texto)

conteo_palabras = (
    rdd_texto2
    .flatMap(lambda x: x.split(" "))
    .map(lambda palabra: (palabra, 1))
    .reduceByKey(lambda a, b: a + b)
)

print(conteo_palabras.collect())
```

```python
# distinct → eliminar duplicados
rdd_distinct = rdd_texto2.flatMap(lambda x: x.split(" ")).distinct()

print("Palabras únicas:", rdd_distinct.collect())


# sortBy → ordenar palabras
rdd_ordenado = rdd_distinct.sortBy(lambda x: x)

print("Palabras ordenadas:", rdd_ordenado.collect())
```

```python
# Agregar acciones estadísticas

rdd_numeros = sc.parallelize([10, 20, 30, 40, 50])

print("Suma:", rdd_numeros.sum())
print("Promedio:", rdd_numeros.mean())
print("Desviación estándar:", rdd_numeros.stdev())
```

```python
# Linaje (DAG)

print(rdd_texto2.toDebugString())
```

## 🔹 Linaje de RDD (DAG)

El linaje de un RDD representa la secuencia de transformaciones aplicadas a los datos.

Spark utiliza este DAG (Directed Acyclic Graph) para reconstruir los datos en caso de fallos y optimizar la ejecución.

Esto permite un procesamiento eficiente y tolerante a errores en entornos distribuidos.

## Conclusión

Se exploraron transformaciones avanzadas de RDD, incluyendo flatMap y reduceByKey, fundamentales para el procesamiento distribuido de datos.

Se implementó un flujo tipo MapReduce, permitiendo transformar, agrupar y agregar información de manera eficiente, sentando las bases para análisis más complejos en entornos Big Data.

# Lección 4: DataFrames y Spark SQL (Nivel Proyecto)

## Introducción

En esta sección se profundiza en el uso de DataFrames para el procesamiento de datos estructurados en Apache Spark.

Se trabajará con operaciones más completas como creación de columnas, transformaciones, filtros avanzados y consultas SQL, simulando un flujo real de análisis de datos en un entorno de retail.

```python


# Crear DataFrame (tipo dataset real)

from pyspark.sql import Row
from pyspark.sql.types import StructType, StructField, StringType, IntegerType

datos = [
    Row(cliente="Ana", producto="Laptop", precio=800, cantidad=1),
    Row(cliente="Luis", producto="Mouse", precio=20, cantidad=2),
    Row(cliente="Pedro", producto="Teclado", precio=50, cantidad=1),
    Row(cliente="Ana", producto="Mouse", precio=20, cantidad=1),
    Row(cliente="Luis", producto="Laptop", precio=800, cantidad=1)
]

# 🔥 Esquema explícito
schema = StructType([
    StructField("cliente", StringType(), True),
    StructField("producto", StringType(), True),
    StructField("precio", IntegerType(), True),
    StructField("cantidad", IntegerType(), True)
])

# 🔥 Crear DataFrame con esquema
df = spark.createDataFrame(datos, schema=schema)

df.show()
df.printSchema()
```

```python
# Crear nuevas columnas 

from pyspark.sql.functions import col

df = df.withColumn("total", col("precio") * col("cantidad"))

df.show()
```

```python
# Filtros más reales

df.filter(col("total") > 100).show()
```

```python
# Agrupación potente

df.groupBy("cliente").sum("total").show()
```

```python
# Ordenar datos

df.orderBy(col("total").desc()).show()
```

```python
# Spark SQL (modo analista)

df.createOrReplaceTempView("ventas")

spark.sql("""
SELECT cliente, SUM(total) as gasto_total
FROM ventas
GROUP BY cliente
ORDER BY gasto_total DESC
""").show()
```

```python
# Caso más real (insight)

spark.sql("""
SELECT producto, COUNT(*) as ventas
FROM ventas
GROUP BY producto
ORDER BY ventas DESC
""").show()
```

```python
# Guardar DataFrame en formato Parquet

# df.write.mode("overwrite").parquet("C:/temp/ventas.parquet")

# print("Archivo guardado en formato Parquet")

df.toPandas().to_parquet("ventas.parquet")

```

## Conclusión

Se aplicaron operaciones avanzadas con DataFrames y Spark SQL, simulando un flujo real de análisis de datos en un contexto de retail.

Estas herramientas permiten transformar, analizar y extraer insights de grandes volúmenes de datos de manera eficiente, siendo fundamentales en entornos de Big Data y analítica empresarial.

# Lección 5: Machine Learning con MLlib

## Introducción

En esta sección se utiliza la biblioteca MLlib de Apache Spark para implementar modelos de Machine Learning sobre datos distribuidos.

Se desarrollará un flujo completo que incluye preparación de datos, creación de features, entrenamiento de modelos y evaluación de resultados, simulando un caso real de analítica en Big Data.

```python
# Dataset (tipo retail)

from pyspark.sql import Row

datos = [
    Row(cliente="Ana", total=800),
    Row(cliente="Luis", total=40),
    Row(cliente="Pedro", total=50),
    Row(cliente="Ana", total=20),
    Row(cliente="Luis", total=800),
    Row(cliente="Pedro", total=30)
]

df = spark.createDataFrame(datos)

df.show()
```

```python
# Crear variable objetivo (clasificación)
# Vamos a clasificar:
# 1 = cliente valioso
# 0 = cliente normal

from pyspark.sql.functions import when

df = df.withColumn(
    "label",
    when(df.total > 100, 1).otherwise(0)
)

df.show()
```

```python
# Feature vector (clave)

from pyspark.ml.feature import VectorAssembler

assembler = VectorAssembler(
    inputCols=["total"],
    outputCol="features"
)

df = assembler.transform(df)

df.show()
```

```python
#Dividir datos

train, test = df.randomSplit([0.7, 0.3], seed=42)
```

```python
# Modelo (Logistic Regression)

from pyspark.ml.classification import LogisticRegression

lr = LogisticRegression(
    featuresCol="features",
    labelCol="label"
)

modelo = lr.fit(train)
```

```python
# Predicciones

predicciones = modelo.transform(test)

predicciones.select("total", "label", "prediction").show()
```

```python
# Evaluación

from pyspark.ml.evaluation import BinaryClassificationEvaluator

evaluador = BinaryClassificationEvaluator(
    labelCol="label"
)

auc = evaluador.evaluate(predicciones)

print("AUC:", auc)
```

## Conclusión

Se implementó un flujo completo de Machine Learning utilizando MLlib en Apache Spark.

Se prepararon los datos, se generaron variables de entrada, se entrenó un modelo de clasificación y se evaluó su desempeño, demostrando la capacidad de Spark para escalar modelos de Machine Learning en entornos Big Data.
